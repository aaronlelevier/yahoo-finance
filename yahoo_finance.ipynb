```python
import collections
import os
from datetime import datetime
import gspread
import requests
from oauth2client.service_account import ServiceAccountCredentials

def main(symbol):
    gc = authenticate()
    for symbol in symbols:
        worksheet = get_worksheet(gc, symbol)
        update_for_symbol(symbol, worksheet)

def authenticate():
    scope = ['https://spreadsheets.google.com/feeds',
             'https://www.googleapis.com/auth/drive']
    filename = '/Users/aaron/stocks-e19fdfa57bd8.json'
    credentials = ServiceAccountCredentials.from_json_keyfile_name(filename, scope)
    gc = gspread.authorize(credentials)
    return gc

def get_worksheet(gc, symbol):
    sh = gc.open('stocks')
    worksheet = sh.worksheet(symbol)
    return worksheet

def update_for_symbol(symbol, worksheet):
    last_insert_date, count = get_last_insert_date(worksheet)
    if last_insert_date != today():
        row = get_summary_data(symbol)
        update_row(worksheet, row, count)
        
def get_last_insert_date(worksheet):
    count = get_row_count(worksheet)
    value = '' if count==0 else worksheet.acell('A%s'%count).value
    return value, count

def get_row_count(worksheet):
    values_list = worksheet.col_values(1)
    return len(values_list)

def today():
    return datetime.now().strftime("%m-%d-%Y")

def get_summary(symbol):
    return "https://query2.finance.yahoo.com/v10/finance/quoteSummary/" + symbol + "?formatted=true&crumb=b.rvJYMbNDw&lang=en-US&region=US&modules=price%2CsummaryDetail%2CpageViews&corsDomain=finance.yahoo.com"

def get_summary_data(symbol):
    summary_url = get_summary(symbol)
    resp2 = requests.get(summary_url)
    data2 = resp2.json()
    row = collections.OrderedDict([
        (u'datetime',today()),
        (u'volume', data2['quoteSummary']['result'][0]['price']['regularMarketVolume']['raw']),
        (u'close', data2['quoteSummary']['result'][0]['summaryDetail']['previousClose']['raw']),
        (u'high', data2['quoteSummary']['result'][0]['summaryDetail']['dayHigh']['raw']),
        (u'open', data2['quoteSummary']['result'][0]['summaryDetail']['open']['raw']),
        (u'low', data2['quoteSummary']['result'][0]['summaryDetail']['dayLow']['raw'])
    ])
    return row

def update_row(worksheet, row, count):
    cells_str = 'A%s:F%s' % (count+1, count+1)
    cells = worksheet.range(cells_str)

    for cell, (k,v) in zip(cells, row.items()):
        cell.value = v

    # Update in batch
    worksheet.update_cells(cells)
    
symbols = ['znga', 'gluu']
main(symbols)
```
